```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movie_data['genres'] = movie_data['genres'].apply(convert)
```

```python
movie_data.head(2)
```

```python
movie_data.head(2)
```

```python
movie_data['keywords'] = movie_data['keywords'].apply(convert)
```

```python
movie_data.head(2)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!= 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movie_data['cast'].apply(convert3)
```

```python
movie_data['cast'] = movie_data['cast'].apply(convert3)
```

```python
movie_data.head(1)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
movie_data['crew'] = movie_data['crew'].apply(fetch_director)
```

```python
movie_data['overview'] = movie_data['overview'].apply(lambda x:x.split())
```

```python
movie_data.head(2)
```

```python
movie_data['genres'] = movie_data['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movie_data['keywords'] = movie_data['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movie_data['cast'] = movie_data['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movie_data['crew'] = movie_data['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movie_data.head(2)
```

```python
movie_data['tags'] = movie_data['overview'] + movie_data['genres'] + movie_data['keywords'] + movie_data['cast'] + movie_data['crew']
```

```python
new_df = movie_data[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances= similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
recommend('The Matrix')
```

```python
recommend('Spectre')
```

```python
recommend("The Dark Knight")
```
